```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred, y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder

# The get_final_recommendation function is modified to be self-contained.
# It now takes the original dataset (df_orig) as an argument to train the
# scaler, encoder, and model on the fly, avoiding FileNotFoundError from .pkl files.
def get_final_recommendation(N, P, K, temperature, humidity, ph, rainfall, original_data_df):
    # Prepare data for training the model within this function call
    X_full = original_data_df.drop('label', axis=1)
    y_full = original_data_df['label']

    # Fit Scaler
    scaler = StandardScaler()
    X_scaled_full = scaler.fit_transform(X_full)

    # Fit LabelEncoder
    label_encoder = LabelEncoder()
    y_encoded_full = label_encoder.fit_transform(y_full)

    # Train a RandomForestClassifier
    # Using similar configuration to previous code (RandomForest, 100 estimators)
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_scaled_full, y_encoded_full)

    # Prepare the single input for prediction
    input_df = pd.DataFrame([[N, P, K, temperature, humidity, ph, rainfall]],
                            columns=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall'])

    # Scale the input using the newly fitted scaler
    input_scaled = scaler.transform(input_df)

    # Make prediction
    prediction_numeric = model.predict(input_scaled)

    # Inverse transform to get the crop name
    prediction_crop_name = label_encoder.inverse_transform(prediction_numeric)

    return prediction_crop_name[0]


# --- TESTING DIFFERENT CROPS (now calling the modified function and passing df_orig) ---

# 1. Test for RICE (High Rainfall is the key differentiator here)
print('--- Testing Rice Conditions ---')
rice_val = get_final_recommendation(N=80, P=40, K=40, temperature=23.0, humidity=80.0, ph=6.5, rainfall=250.0, original_data_df=df_orig)
print(f'Recommended: {rice_val}')

# 2. Test for WHEAT (Higher K and different rainfall/temp)
print('\n--- Testing Wheat Conditions ---')
wheat_val = get_final_recommendation(N=70, P=80, K=130, temperature=25.0, humidity=60.0, ph=6.5, rainfall=180.0, original_data_df=df_orig)
print(f'Recommended: {wheat_val}')
```
