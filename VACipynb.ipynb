```python
orders = pd.read_csv("/content/olist_order_reviews_dataset.csv")
merged_df = pd.merge(df1, orders, on="order_id", how="inner")
```

```python
merged_df['order_id'].fillna("Unknown", inplace=True)
```

```python
df = merged_df

print("\nMerged Dataset Shape :", df.shape)
print(df.head())
```

```python
df.drop_duplicates(inplace=True)

# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna("Missing", inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

print("\nMissing Values After Cleaning:\n", df.isnull().sum().sum())
```

```python
date_cols = [col for col in df.columns if "timestamp" in col]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Extract Year / Month
if "order_purchase_timestamp" in df.columns:
    df["purchase_year"] = df["order_purchase_timestamp"].dt.year
    df["purchase_month"] = df["order_purchase_timestamp"].dt.month

# Standardize city names
if "customer_city" in df.columns:
    df["customer_city"] = df["customer_city"].str.title()
```

```python
# Distribution of Review Scores
if "review_score" in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(x="review_score", data=df)
    plt.title("Distribution of Review Scores")
    plt.show()
```

*This graph shows which states have the highest number of customers. States with taller bars contribute more orders/customers, indicating stronger market demand in those regions. Lower bars represent states where customer presence is smaller.*

```python

# Payment Value Histogram
if "payment_value" in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df["payment_value"], bins=30, kde=True)
    plt.title("Distribution of Payment Value")
    plt.show()
```

*This plot explains how customer zip codes are spread across the dataset. Peaks in the histogram show areas where many customers belong to similar zip code ranges, while gaps indicate fewer customers in those locations. It helps identify concentration of customers geographically.*

```python
# Payment by Review Score
if "review_score" in df.columns and "payment_value" in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x="review_score", y="payment_value", data=df)
    plt.title("Payment Value vs Review Score")
    plt.show()
```

*This graph compares zip code ranges for each state. It shows the median, spread, and outliers of zip codes within every state, helping understand regional variability. States with wider boxes have more diverse customer locations.*

```python
top_cities = merged_df['order_id'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title("Top 10 Customer Cities")
plt.show()
```

*This plot highlights the cities with the highest number of customers. Cities with longer bars are the most active and valuable markets for business growth. It helps identify urban areas where customer demand is strongest.*

*The heatmap shows relationships between numeric variables. Values close to 1 mean strong positive correlation, close to -1 mean negative correlation, and near 0 mean weak/no relationship. This helps in feature selection and understanding data dependencies.*

```python
if "payment_type" in df.columns:
    pay = df["payment_type"].value_counts()

    plt.figure(figsize=(8,8))
    plt.pie(pay.values, labels=pay.index, autopct="%1.1f%%", startangle=90)
    plt.title("Payment Type Distribution")
    plt.show()
```

*This chart shows the percentage contribution of the top five states in total customers. Bigger slices represent states with larger customer shares. It is useful for understanding market dominance and customer distribution proportionally.*

```python
if "purchase_month" in df.columns:
    monthly = df["purchase_month"].value_counts().sort_index()

    plt.figure(figsize=(10,5))
    sns.lineplot(x=monthly.index, y=monthly.values, marker="o")
    plt.title("Monthly Orders Trend")
    plt.xlabel("Month")
    plt.ylabel("Orders")
    plt.show()
```

*This scatter plot visualizes how zip codes vary across different states using colors. Clusters of points indicate states with concentrated customer regions, while scattered points show wider geographical spread. It helps detect patterns and overlaps among states.*

```python
df_encoded = df.copy()
le = LabelEncoder()

for col in df_encoded.columns:
    if df_encoded[col].dtype == "object":
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

# Convert datetime to numeric
for col in df_encoded.select_dtypes(include=["datetime64[ns]"]).columns:
    df_encoded[col] = df_encoded[col].astype("int64") // 10**9

# Correlation Matrix
corr = df_encoded.corr()

plt.figure(figsize=(18,12))
sns.heatmap(corr, cmap="coolwarm", annot=False, linewidths=0.2)
plt.title("Correlation Matrix - Merged Dataset")
plt.show()
```
